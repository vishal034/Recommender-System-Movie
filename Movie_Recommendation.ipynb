```python
import numpy as np
import pandas as pd 
movies=pd.read_csv('tmdb_5000_movies.csv',encoding='utf-8')
credits=pd.read_csv('tmdb_5000_credits.csv',encoding='utf-8')
movies=movies.merge(credits,on='title')
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
import ast
ast.literal_eval(movies.iloc[0].genres)
```

```python
def convert(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=5:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
import json

def fetch_dir(obj):
    l = []
    print("Input string:", obj)  # Debugging print statement
    if isinstance(obj, str):
        # Convert JSON string to Python object
        crew_list = json.loads(obj)
        print("Parsed list:", crew_list)  # Debugging print statement
    else:
        # Assume obj is already a Python object
        crew_list = obj
    
    # Check if crew_list is a list
    if isinstance(crew_list, list):
        for member in crew_list:
            # Ensure member is a dictionary before accessing its keys
            if isinstance(member, dict) and member.get('job') == 'Director':
                l.append(member.get('name'))
    return l

```

```python
# Apply the fetch_dir function to the 'crew' column
movies['crew'] =movies['crew'].apply(fetch_dir)
```

```python
movies.head(2)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
import pandas as pd
import numpy as np
import nltk
```

```python
movies=pd.read_csv('movies.csv')
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return" ".join(y)
```

```python
ps.stem('loved')
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index=movies[movies['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print (movies.iloc[i[0]].title)
       
```

```python
recommend('Superman')
```
